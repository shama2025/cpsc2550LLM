```python
from gensim.models import Word2Vec
import gensim.downloader

vectors = gensim.downloader.load('glove-wiki-gigaword-50')
```

```python
vectors['jesus']
```

```python
len(vectors['jesus'])
```

```python
vectors.similarity('satan', 'jesus') #On a range the output as closer it is to 1 then the more similar the words are 
```

```python
vectors.similar_by_vector(vectors['jesus']) #Top 10 of the most similar words
```

```python
len(vectors.similar_by_vector(vectors['jesus'])) 
```

```python
import math

def vlen(vec):
    return math.sqrt(sum(c*c for c in vec))

vlen(vectors['played'] - vectors['play'])
vlen(vectors['play'])
```
